```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '../data/coqa_train.csv'
coqa_data = pd.read_csv(file_path)


display(coqa_data.head())


source_counts = coqa_data['source'].value_counts()
coqa_data['num_questions'] = coqa_data['questions'].apply(lambda x: len(eval(x)))
average_questions_per_story = coqa_data['num_questions'].mean()


plt.figure(figsize=(10, 6))
source_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Entries by Source')
plt.xlabel('Source')
plt.ylabel('Number of Entries')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(coqa_data['num_questions'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Number of Questions per Story')
plt.xlabel('Number of Questions')
plt.ylabel('Frequency')
plt.show()


print(f'Average number of questions per story: {average_questions_per_story}')
```
